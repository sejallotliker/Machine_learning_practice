#Building pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
columns= X_train_df.select_dtypes(exclude=[np.number]).columns.to_list()
columns
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
#Since there is no numerical data in train as well as test, we only need to do non-numerical imputation.
nonnumerical_preprocessor = Pipeline([
    ('impute', SimpleImputer(strategy="most_frequent")),
    ('encoder', OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])
```

```python
type(nonnumerical_preprocessor), isinstance(nonnumerical_preprocessor, Pipeline), nonnumerical_preprocessor
```

```python
X_preprocessed = nonnumerical_preprocessor.fit_transform(X_train_df)
```

```python
from sklearn.linear_model import LinearRegression
full_pipeline = Pipeline([
    ('preprocessor', nonnumerical_preprocessor),
    ('linear_regression', LinearRegression())
])
```

```python
full_pipeline.fit(X_train_df, y_train_df)
```

```python
yhat_test = full_pipeline.predict(X_test_df)
```

```python
#rmse
from sklearn.metrics import root_mean_squared_error as rmse
lr_test_rmse = rmse(y_test_df, yhat_test)
lr_test_rmse
```

```python
yhat_train = full_pipeline.predict(X_train_df)
```

```python
#rmse
from sklearn.metrics import root_mean_squared_error as rmse
lr_train_rmse = rmse(y_train_df, yhat_train)
lr_train_rmse
```

```python
#R2 score:

from sklearn.metrics import r2_score
lr_test_r2 = r2_score(y_test_df, yhat_test)
lr_test_r2
```

```python
lr_train_r2 = r2_score(y_train_df, yhat_train)
lr_train_r2
```
