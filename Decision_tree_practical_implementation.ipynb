```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

##We can import iris data from other way

```python
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
```

```python
#independent and dependent features
X = df.iloc[:,:-1]
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
#To use decision tree
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train,y_train)
```

```python
y_hat = tree_model.predict(X_test)
```

```python
from sklearn import tree
plt.figure(figsize = (15, 10))
tree.plot_tree(tree_model,filled=True) #The boxes in the tree will be coloured and filled with info
```

```python
#Note that gini index at leaf node is always zero
```

```python
#We can use the hyperparameter as max depth
tree_model_2 = DecisionTreeClassifier(max_depth=2)
tree_model_2.fit(X_train,y_train)
```

```python
y_hat_2 = tree_model_2.predict(X_test)
```

```python
from sklearn import tree
plt.figure(figsize = (15, 10))
tree.plot_tree(tree_model_2,filled=True) #The boxes in the tree will be coloured and filled with info
```

```python
#To see the accuracy of the models

from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_test,y_hat))
```

```python
print(classification_report(y_test,y_hat))
```

```python
print(accuracy_score(y_test,y_hat_2))
```

```python
print(classification_report(y_test,y_hat_2))
```

####Therefore, we see that on this dataset , we do not see much of a difference in accuracy after pruning and before pruning
